```python
import pandas as pd
from anndata import AnnData

#筛选control组和PFC的细胞标签
control_df = df[df['diagnosis'] == 'Control']
PFC_control_df = control_df[control_df['region'] == 'PFC']
BA24_PFC_control_df = PFC_control_df[PFC_control_df['sample'] == '5408_PFC_Nova']

# 创建一个字典，将cellId映射到cell_type
CellId_to_cell_type = dict(zip(PFC_control_df['cell'], PFC_control_df['cluster']))

# 创建一个新的AnnData对象，仅包含匹配成功的obs
adata_matched = adata[adata.obs_names.isin(CellId_to_cell_type)]

# 将映射结果应用于adata_matched的obs列
adata_matched.obs['cell_type'] = [CellId_to_cell_type[obs_name] for obs_name in adata_matched.obs_names]
# 现在adata_matched包含了匹配成功的obs，并且obs['cell_type']列已经包含了匹配成功的cell_type值。
```
