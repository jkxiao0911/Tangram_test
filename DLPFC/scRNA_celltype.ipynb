```python
import anndata
import h5py
import scipy.sparse as sp

# 1. 打开.hdf5文件
hdf5_file_path = '/mnt/disk1/xiaojk/data/DLPFC/GSM5138547_RL2128_20y_b_snRNAseq_filtered_feature_bc_matrix.h5'
hdf5_file = h5py.File(hdf5_file_path, 'r')

indices = hdf5_file['matrix']['indices'][:]
indptr = hdf5_file['matrix']['indptr'][:]
data = hdf5_file['matrix'] ['data'][:]

#创建CSR格式的稀疏矩阵
sparse_matrix_shape = (hdf5_file['matrix'] ['barcodes'].shape[0],hdf5_file['matrix'] ['features']['name'].shape[0])
sparse_matrix = sp.csr_matrix((data, indices, indptr), shape=sparse_matrix_shape)
adata = anndata.AnnData(X=sparse_matrix)

# 将字节字符串转换为普通字符串并去掉'b'
adata.obs_names = hdf5_file['matrix']['barcodes'][:].astype(str)
adata.obs_names = ["RL2085#" + name for name in adata.obs_names]
adata.var_names = hdf5_file['matrix']['features']['name'][:].astype(str)

adata.var['gene_id'] = hdf5_file['matrix']['features']['id'][:].astype(str)
```

```python
df = pd.read_csv('/mnt/disk1/xiaojk/data/DLPFC/Velmeshev_2019/meta.txt',sep = '\t')
```

```python
adata_matched = ad_sc.copy()
```

```python
#删除cell名中多余的字符串
adata_obs_names = ad_sc.obs_names.tolist()
adata_obs_names = [obs_name.split('#')[1] for obs_name in adata_obs_names]
adata_obs_names = pd.Index(adata_obs_names)
adata_matched.obs_names = adata_obs_names
adata_matched.obs_names
```

```python
# 创建一个字典，将cellId映射到cell_type
cell_list = df['cell'].tolist()
cell_list = [cell.split('_')[0] for cell in cell_list]
CellId_to_cell_type = dict(zip(cell_list, df['cluster']))
CellId_to_cell_type
```

```python
# 创建一个新的AnnData对象，仅包含匹配成功的obs
adata_matched = adata_matched[adata_obs_names.isin(CellId_to_cell_type)]
adata_matched.obs['cell_type'] = [CellId_to_cell_type[obs_name] for obs_name in adata_matched.obs_names]
```
